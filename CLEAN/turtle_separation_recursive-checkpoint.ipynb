```python
import re

def extract_triples_with_relation_recursive(file_path, source, relations, visited=None):
    if visited is None:
        visited = set()

    prohibited_strings = ['Permen_Dagri', 'Permen_Desa', 'Permen_Dikbud', 'Permen_Kominfo', 'Permen_PANRB', 'Permen_PPPA', 'Permen_Par', 'Permen_Ristekdikti']
    triples = []
    with open(file_path, 'r') as file:
        for line in file:
            for relation in relations:
                relation = 'lexid-s:' + relation
                match = re.match(fr'{source}\s*{relation}\s*lexid:([^ ]+)', line)
                if match:
                    target = 'lexid:' + match.groups()[0]
                    if all(prohibited not in target for prohibited in prohibited_strings):
                        if target not in visited:
                            visited.add(target)
                            triples.append((source, relation, target))
                            print(f"{source} {relation} {target} .")
                            triples.extend(extract_triples_with_relation_recursive(file_path, target, relations, visited))
    return triples

file_path = 'pp4-2019bt.ttl'
source = 'lexid:PP_2019_4'
relations = ['considers', 'dct:description', 'name', 'owl:sameAs', 'hasCreator', 'hasDictum', 'hasCreator', 'hasEnactionDate', 'hasEnactionLocation', 'hasEnactionOffice', 'hasEnactionOfficial', 'hasPromulgationDate', 'hasPromulgationLocation', 'hasPromulgationOffice', 'hasPromulgationOfficial', 'hasPromulgationPlace', 'hasRegulationNumber', 'hasRegulationYear', 'rdfs:label', 'amendedBy', 'amends', 'implementedBy', 'implements', 'hasLegalBasis', 'legalBasisOf', 'repealedBy', 'repeals', 'hasContent', 'hasPart', 'isContentOf', 'isPartOf', 'adds', 'hasAdditionContent', 'hasAdditionTarget', 'deletes', 'modifies', 'hasModificationContent', 'hasAct', 'hasActType', 'hasElement', 'hasCondition', 'hasModality',  'hasObject', 'hasQualifier', 'hasQualifierType', 'hasQualifierValue', 'hasRule', 'hasSubject', 'isRuleOf', 'refersTo']
triples = extract_triples_with_relation_recursive(file_path, source, relations)
for i in triples:
  print(i)
```
