```python
# Peraturan --> Peraturan

dff = df[df['label']=='Peraturan'].drop_duplicates(subset='code')
dff['formalTyping'] = dff['formalTyping'].apply(lambda x: x.replace('Peraturan ', ''))
dff['code'] = dff['code'].apply(lambda x: x.replace('Peraturan_', ''))

tuples_list = list(zip(dff['formalTyping'], dff['code']))
for i in tuples_list:
    print(i)
```
