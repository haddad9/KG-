````python
generation_config_sl = model_sl.generation_config
generation_config_sl.temperature = 0.0
generation_config_sl.max_new_tokens = 500
generation_config_sl.use_cache = False
generation_config_sl.repetition_penalty = 1.5
generation_config_sl.do_sample = False
generation_config_sl.top = 0.7
generation_config_sl.pad_token_id = tokenizer_sl.eos_token_id
generation_config_sl.eos_token_id = tokenizer_sl.eos_token_id
generation_config_sl
````

````python
prompt = """
The following is a friendly conversation between a human and an AI. The AI is talkative and provides lots of specific details from its context.

Current conversation:

Human: What is x value that fit the linear equation: 2x+3=7?
AI:
""".strip()
input_ids = tokenizer_sl(prompt, return_tensors="pt").input_ids
input_ids = input_ids.to(model_sl.device)
````

````python
%%time

with torch.inference_mode():
    outputs = model_sl.generate(
        input_ids = input_ids,
        generation_config = generation_config_sl,
    )
````

````python
response = tokenizer_sl.decode(outputs[0], skip_special_tokens=True)
print(response)
````
