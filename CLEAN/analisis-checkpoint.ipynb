```python
from rdflib import Graph, Namespace, Literal

def literals_case_insensitive_equal(lit1, lit2):
    return str(lit1).lower() == str(lit2).lower()

def triple_case_insensitive_equal(triple1, triple2):
    return (
        triple1[0] == triple2[0] and
        triple1[1] == triple2[1] and
        (
            (isinstance(triple1[2], Literal) and isinstance(triple2[2], Literal) and literals_case_insensitive_equal(triple1[2], triple2[2])) or
            (not isinstance(triple1[2], Literal) and not isinstance(triple2[2], Literal) and triple1[2] == triple2[2])
        )
    )

def add_prefixes(graph, prefix_str):
    for line in prefix_str.strip().split('\n'):
        if line.startswith('@prefix'):
            parts = line.split()
            prefix = parts[1][:-1]
            uri = parts[2][1:-1]
            graph.namespace_manager.bind(prefix, URIRef(uri))

def triple_to_prefixed_str(triple, graph):
    return ' '.join(graph.namespace_manager.normalizeUri(part) for part in triple)

def triple_case_insensitive_equal(triple1, triple2):
    return all(str(part1).lower() == str(part2).lower() for part1, part2 in zip(triple1, triple2))


def evaluate_ttl(ttl_pred, ttl_gold):
    g_pred = Graph()
    g_pred.parse(data=ttl_pred, format="turtle")

    g_gold = Graph()
    g_gold.parse(data=ttl_gold, format="turtle")

    g_pred_total = len(g_pred)
    g_gold_total = len(g_gold)

    tp = len(set(g_pred).intersection(set(g_gold)))

    fp = g_pred_total - tp
    fn = g_gold_total - tp
    
    if tp + fp == 0:
        p = 0
    else:
        p = round(tp / (tp + fp), 3)
        
    if tp + fn == 0:
        r = 0
    else:
        r = round(tp / (tp + fn), 3)
    
    if p + r == 0:
        f1 = 0
    else:
        f1 = round((2 * p * r)/ (p + r), 3)
    
    return tp, fp, fn, p, r, f1

def evaluate_ttl_case_insensitive(ttl_pred, ttl_gold, prefix_str=prefix):
    g_pred = Graph()
    g_pred.parse(data=ttl_pred, format="turtle")

    g_gold = Graph()
    g_gold.parse(data=ttl_gold, format="turtle")

    add_prefixes(g_pred, prefix_str)
    add_prefixes(g_gold, prefix_str)

    g_pred_total = len(g_pred)
    g_gold_total = len(g_gold)

    tp = 0
    fp_triples = []
    fn_triples = []

    for triple in g_pred:
        if any(triple_case_insensitive_equal(triple, gold_triple) for gold_triple in g_gold):
            tp += 1
        else:
            fp_triples.append(triple_to_prefixed_str(triple, g_pred))

    for triple in g_gold:
        if not any(triple_case_insensitive_equal(triple, pred_triple) for pred_triple in g_pred):
            fn_triples.append(triple_to_prefixed_str(triple, g_gold))

    fp = g_pred_total - tp
    fn = g_gold_total - tp

    if tp + fp == 0:
        p = 0
    else:
        p = round(tp / (tp + fp), 3)
        
    if tp + fn == 0:
        r = 0
    else:
        r = round(tp / (tp + fn), 3)
    
    if p + r == 0:
        f1 = 0
    else:
        f1 = round((2 * p * r) / (p + r), 3)

    print("False Positives (FP):")
    for triple in fp_triples:
        print(triple)

    print("\nFalse Negatives (FN):")
    for triple in fn_triples:
        print(triple)
    
    return tp, fp, fn, p, r, f1

prefix = '''@prefix xsd: <http://www.w3.org/2001/XMLSchema#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix dbo: <http://dbpedia.org/ontology/> .
@prefix dct: <http://purl.org/dc/terms/> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix wd: <https://www.wikidata.org/wiki/> .
@prefix lexid-s: <https://w3id.org/lex-id/schema/> .
@prefix lexid: <https://w3id.org/lex-id/data/> .

'''
```

```python
from rdflib import Graph

def print_triples_with_prefix(triples, graph):
    for triple in triples:
        s, p, o = triple
        print(graph.qname(s), graph.qname(p), graph.qname(o))

def check_triples(ttl_pred, ttl_gold):
    graph1 = Graph()
    graph2 = Graph()

    # Memuat data RDF ke dalam graf (ganti 'file1.rdf' dan 'file2.rdf' dengan file RDF anda)
    graph1.parse(data=ttl_pred, format="turtle")
    graph2.parse(data=ttl_gold, format="turtle")
    set1 = set(graph1)
    set2 = set(graph2)
    # Mengidentifikasi triples
    intersection = set1 & set2
    only_in_graph1 = set1 - set2
    only_in_graph2 = set2 - set1

    # Fungsi untuk mencetak triples dengan prefiks
    

    # Mencetak hasil
    print("Triples yang ada di kedua graf:")
    print_triples_with_prefix(intersection, graph1)

    print("\nTriples yang hanya ada di graf pertama:")
    print_triples_with_prefix(only_in_graph1, graph1)

    print("\nTriples yang hanya ada di graf kedua:")
    print_triples_with_prefix(only_in_graph2, graph2)
```
